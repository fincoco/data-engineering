__Implement your own Circle class, which inherits Point class.__ It should inherit/override a function `distance` such that distance can be calculated between both points and circles.  The `distance` function in `Circle` class should calculate the distance between the circle's center and the given point / the given circle's center.  
_x and y member variable will be a location of circle's center in Circle class._

```python
# Write your own code!
from math import sqrt

class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def print_point(self):
    print("<", self.x, ",", self.y, ">")

  def distance(self, other):
    return sqrt((self.x - other.x) ** 2 + (self.y - other.y) ** 2)   

class Circle(Point):
    def __init__(self, center, radius):
        super().__init__(center[0], center[1])
        self.radius = radius

    def print_circle(self):
        self.print_point()
        print("radius:", self.radius)

    def in_circle(self, point):
        return self.distance(point) < self.radius
```

```python
from math import sqrt

class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def print_point(self):
    print("<", self.x, ",", self.y, ">")

  def distance(self, other):
    return sqrt((self.x - other.x) ** 2 + (self.y - other.y) ** 2)   

class Circle(Point):
  def __init__(self, center, radius):
    super(Circle, self).__init__(center[0], center[1])
    self.radius = radius

  def in_circle(self, point):
    return self.distance(point) < self.radius
```

```python
# super(클래스, self): 상속받는 부모 클래스를 받아온다
#   > Circle이 Point 클래스를 받아온다
super().__init__()
super(Circle, self).__init__()
```

```python
class A:
    def __init__(self):
        self.a = 10

    def get_a(self):
        return self.a

class B(A):
    def __init__(self):
        super(B,self).__init__()
        self.b = 20

    def get_b(self):
        return self.b

class C(B):
    def __init__(self):
        super(B,self).__init__() #B의 부모클래스인 A를 상속받기 때문에 get_b() 사용할 수 없음
        self.c = 30

    def get_c(self):
        return self.c

new_c = C()
print(new_c.get_a())
print(new_c.get_c())
print(new_c.get_b())
```
