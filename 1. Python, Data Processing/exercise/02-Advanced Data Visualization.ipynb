### 7. Read the total profit of each month and show it using the histogram to see most common profit ranges
The histogram should look like this.
![ex7](https://drive.google.com/uc?export=download&id=1-9sbbEFvrNVRNQAeYiU-Vbc9AMiQcu3E)

```python
# Write your own code!
# profit range에 따라 count 만들기

print(profitList)

count = np.zeros(9)
for x in profitList:
  if x >= 150000 and x < 175000: count[0] += 1
  elif x >= 175000 and x < 200000: count[1] += 1
  elif x >= 200000 and x < 225000: count[2] += 1
  elif x >= 225000 and x < 250000: count[3] += 1
  elif x >= 250000 and x < 275000: count[4] += 1
  elif x >= 275000 and x < 300000: count[5] += 1
  elif x >= 300000 and x < 325000: count[6] += 1
  elif x >= 325000 and x < 350000: count[7] += 1
  else: count[8] += 1
print(count)

count = np.zeros(9)
bins = [150000, 175000, 200000, 225000, 250000, 275000, 300000, 325000, 350000]
range_count = np.digitize(profitList, bins) # 해당 구간에 속하는 숫자들의 개수를 리턴함
print(range_count)

for i in range_count:
  count[i-1] += 1

print(count)
```

```python
# 히스토그램은 자동으로 y축을 도수로 설정하기 때문에 따로 count할 필요 없음 
plt.hist(
  profitList, range = (150000, 350000), bins = 8,  # range로 구간 설정, bin으로 막대 개수 설정
  label = "Profit data"
)

plt.xlabel("Profit range in dollar")
plt.ylabel("Count")
plt.legend(loc = "upper left")
plt.show()
```

### 8. Calculate total sale data for last year for each product and show it using a Pie chart

Note: In Pie chart display Number of units sold per year for each product in percentage.

The Pie chart should look like this.

![ex8](https://drive.google.com/uc?export=download&id=1-B3ZjiGyLY33j3CnWnT1zCb0pDwfutXQ)


```python
# Write your own code!
count = [0 for i in range(6)]

datas = {
    "Face cream": faceCreamSalesData,
    "Face wash": faceWashSalesData,
    "Toothpaste": toothPasteSalesData,
    "Bathing soap": bathingsoapSalesData,
    "Shampoo": shampooSalesData,
    "Moisturizer": moisturizerSalesData,
}

i = 0
for label, val in datas.items():
  count[i] = np.sum(datas[label])
  i += 1

plt.pie(
    count, labels = datas.keys(), autopct ='%1.1f%%'
)
plt.legend(loc = 'lower right')
plt.title("Sales Data")
plt.show()
```
