### P2. Basic bar charts

1. Generate a list of size N, filled with records of N dice rolls (add 1 if dice showed side 1).  
2. save the values in record[]

** for generating dice roll, use random() function **
- https://wikidocs.net/79

```python
# Your code here
from random import randint

# 10000번 실험할 때, 1번 실험시 주사위 100번 던져서 1이 몇 번 나오는가? 
records = []
for i in range(10000):
  record = 0
  for j in range(100):
    if randint(1, 6) == 1: 
      record += 1
  records.append(record)
  
print(records)
```

Run code below to change records to probability

```python
# Your code here
count = [0] * 101

for r in records:
  count[r] += 1

print(count)
```

Now draw bar chart for probability, refer to https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.bar.html  
Now draw plot for normal distribution, refer to https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.plot.html
- Requirements for plot : linewidth = 2, color = 'y' 
- Equation for normal distribution : 
$$
y = \frac{1}{\sqrt{2\pi \sigma}} e^{\frac{-\left ( x-\mu \right )^{2}}{2\sigma^{2}}}
$$

```python
# Your code here
probability = [i / 10000 for i in count]
print(probability)
```

### draw bar chart and plot normal distribution code

~~~ python
x = np.arange(len(probability))
plt.bar(x, probability)
plt.plot(x, 1/(sigma * np.sqrt(2 * np.pi)) *np.exp( - (x - mu)**2 / (2 * sigma**2) )   ,  linewidth = 2, color = 'y')
plt.show()
~~~

```python
import numpy as np
from math import sqrt

mu, sigma = 100 * (1/6), np.sqrt(100* (1/6) * (5/6))
plt.plot(x, 1/(sigma * np.sqrt(2 * np.pi)) *np.exp( - (x - mu)**2 / (2 * sigma**2) )   ,  linewidth = 2, color = 'y') # 정규분포 구하는 식 
plt.show()
```
