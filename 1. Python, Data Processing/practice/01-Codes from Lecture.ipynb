**p62** Notes
~~~python
x = lambda a, b : a * b
print(x(5,6))

def myfunc(n):
    return lambda a : a*n

mydoubler = myfunc(2)
print(mydoubler(11))
~~~

```python
# Lambda 함수를 쓰는 이유:
# 1. 일단 일차적으로 편함
# 2. 메모리의 절약
#    - 기본적으로 def를 통해 함수를 선언한다. 코드와 이름을 담아서 함수 클래스를 통해 객체를 생성하고, 그 객체를 함수 이름과 동일한 변수에 담아 준다 (def가 해줌).
#    - 변수에 담는다는 것은 기본적으로 메모리를 사용한다는 것을 의미
#    - 반면 lambda 함수는 함수의 이름을 지정하지 않는다. 따라서 저장된 변수가 없고, 메모리를 잡아먹지 않는다.
#    - 이 경우 변수에 담은 것이긴 함

func = lambda a, b : a * b
print( func(5,6) )

def myfunc(n):
    return lambda a : a * n
# myfunc는 n을 받아, a * n을 return하는 함수를 return하는 함수

mydoubler = myfunc(2)
# a * 2를 return하는 함수를 return하여 mydoubler 내부에 넣어줬다

print( mydoubler(11) )
```

**p76** Example : Molecule Class
~~~python
class molecule:
    def __init__(self, name = 'Generic'):
        self.name = name
        self.atomlist = []
    def addatom(self, atom):
        self.atomlist.append(atom)
    def __repr__(self):
        str = 'This is a molecule named %s\n' % self.name

        for atom in self.atomlist:
            str = str + repr(atom) + '\n'
        return str

mol = molecule('Water')
at = atom(8, 0, 0, 0)
~~~

```python
class molecule:
    def __init__(self, name = 'Generic'):
        self.name = name
        self.atomlist = []

    def addatom(self, atom):
        self.atomlist.append(atom)
        
    def __repr__(self):
        string = f'This is a molecule named {self.name}\n'
        string = string + 'It has %d atoms \n' % len(self.atomlist)
        for atom in self.atomlist:
            string = string + repr(atom) + '\n'
        return string

mol = molecule('Water')
at = atom(8, 0, 0, 0)
mol.addatom(at)
mol.addatom(atom(1, 0, 0, 1))
mol.addatom(atom(1, 0, 1, 0))
print(mol.name)
print()
print(mol)
```

```python
class monoatomic_molecule(molecule):
    def num_of_molecule(self):
        return len(self.atomlist)

    def addatom(self, atom):
        if len(self.atomlist) == 0:
            super(monoatomic_molecule, self).addatom(atom)
        else:
            for at in self.atomlist:
                if at.atno != atom.atno:
                    print("This molecule is monoatomic!")
                    return
            super(monoatomic_molecule, self).addatom(atom)

monomol = monoatomic_molecule('Ozone')
monomol.addatom(atom(8, 1, 0, 0))
monomol.addatom(atom(8, 0, 1, 0))
monomol.addatom(atom(8, 0, 0, 1))
print(monomol)

monomol.addatom(atom(2, 0, 0, 1)) # at.atno(8)과 atom.atno(2)가 다르므로 add되지 않고 retrun 
print(monomol)
```
