### Point Class

Now, let's define a Point class. It represents a point in 2-dimnesional space.  
`Point` class has constructor `__init__(self, x, y)` which specifies the point's location. `print_point(self)` prints its location. `distance(self, other)` returns distance between 2 points.


~~~ python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def print_point(self):
        print ("<", self.x, "," ,self.y, ">")

    def distance(self, other):
        return sqrt((self.x - other.x) ** 2 + (self.y - other.y) ** 2)
~~~

```python
### YOUR CODE HERE ###
class Point:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def print_point(self):
    print("<", self.x, ",", self.y, ">")

  def distance(self, other):
    return sqrt((self.x - other.x) ** 2 + (self.y - other.y) ** 2)   
```

### Circle Class

Circle class represents a circle in 2-dimensional space.  
It has `self.center` and `self.radius` as member variable, and `self.center` is an instance of Point class we defined above.  

~~~ python
class Circle:
    def __init__(self, center, radius):
        self.center = Point(center[0], center[1])
        self.radius = radius

    def print_circle(self):
        print ("center:",self.center.x, self.center.y ,", radius:", self.radius)

    def in_circle(self, point):
        return self.center.distance(point) < self.radius
~~~

```python
### YOUR CODE HERE ###
class Circle:
    def __init__(self, center, radius):
        self.center = Point(center[0], center[1])
        self.radius = radius

    def print_circle(self):
        print ("center:",self.center.x, self.center.y ,", radius:", self.radius)

    def in_circle(self, point):
        return self.center.distance(point) < self.radius
```
